```python
%matplotlib nbagg

import cv2

import numpy as np
from matplotlib import pyplot as plt

capture = cv2.VideoCapture(0)
n=0
plt.ion()

while(True):
    ret, frame = capture.read()
    mean = frame.mean()
    plt.plot(n, mean, marker="o", color="red")
    plt.show
    print('mean',mean)
    cv2.imshow('frame',frame)
    n=n+1
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

capture.release()
cv2.destroyAllWindows()
```
